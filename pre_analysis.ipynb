# Heart Disease Data Analysis

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/cleveland_data.csv')

df.head()
```

## Data Description

•    age: The person's age in years

•    sex: The person's sex (1 = male, 0 = female)

•    cp: The chest pain experienced (Value 1: typical angina, Value 2: atypical angina, Value 3: non-anginal pain, Value 4: asymptomatic)

•    trestbps: The person's resting blood pressure (mm Hg on admission to the hospital)

•    chol: The person's cholesterol measurement in mg/dl

•    fbs: The person's fasting blood sugar (> 120 mg/dl, 1 = true; 0 = false)

•    restecg: Resting electrocardiographic measurement (0 = normal, 1 = having ST-T wave abnormality, 2 = showing probable or definite left ventricular hypertrophy by Estes' criteria)

•    thalach: The person's maximum heart rate achieved

•    exang: Exercise induced angina (1 = yes; 0 = no)

•    oldpeak: ST depression induced by exercise relative to rest ('ST' relates to positions on the ECG plot. See more here)\n
•    slope: the slope of the peak exercise ST segment (Value 1: upsloping, Value 2: flat, Value 3: downsloping)

•    ca: The number of major vessels (0-3)

•    thal: A blood disorder called thalassemia (3 = normal; 6 = fixed defect; 7 = reversable defect)

•    target: Heart disease (0 = no, 1+ = yes)

## Null Value Check

```python
plt.figure(figsize=(20,10))
plt.xticks(size=20,color='grey')
plt.tick_params(size=12,color='grey')
plt.title('Null Value Heatmap',color='grey',size=30)
sns.heatmap(df.isnull(),
            yticklabels=False,
            cbar=False,
            cmap='PuBu_r',
            )
```

## Correlation Matrix

```python
plt.figure(figsize=(14,10))
sns.heatmap(df.corr(),annot=True,cmap="magma",fmt='.2f')

plt.show()
```

```python
sns.countplot(x ='sex', hue = "target", data = df)

plt.show()
```

```python
dfcount = df.copy()

def sexlbl(sex):
    if sex == 0:
        return 'female'
    else:
        return 'male'
    
def cadlbl(target):
    if target == 0:
        return 'No CAD'
    else:
        return 'CAD'
    
dfcount['sex'] = dfcount['sex'].apply(sexlbl)
dfcount['target'] = dfcount['target'].apply(cadlbl)
```

```python
sns.countplot(x ='sex', hue = "target", data = dfcount)

plt.show()
```
